Index the CSV by Title, then change the labels of the columns

```python
df1 = pd.read_csv("IMDB-Movie-Data.csv", index_col="Title")
```

```python
df1.columns = ["rank", "genre", "desc", "dir", "actor", "year", "length", "rating", "votes", "rev", "mscore"]
```

```python
df1.index.name = "title"
```

```python
df1.info()
```

```python
df1.shape
```

```python
df1.describe()
```

find what entries have a null value:

```python
df1.isnull().sum()
```

Revenues and Metascore have null values, they will now be changed:

```python
df1['rev'].fillna(df1['rev'].mean(),inplace=True)
```

```python
df1['mscore'].fillna(df1['mscore'].mean(),inplace=True)
```

Verify that the rev and mscore columns no longer have null values:

```python
df1.isnull().sum()
```

Drop the rank column:

```python
df1.drop(['rank'],axis=1,inplace=True)
```

Note: axis=0 is for rows, axis=1 is for columns

```python
df1.groupby(["year"]).mean()
```

```python
df1.groupby(["year"]).mean().sort_values(by=["rev"],ascending=False)
```

Sort the movies from 2012 by revenue:

```python
df1[df1['year'] == 2012].sort_values(by=['rev'],ascending=False)
```

```python
df1[df1['year'] == 2009].sort_values(by=['rev'],ascending=False)
```

What were the highest rated movies in the Comedy and Action genres?

```python
df1[df1['genre'].isin(['Comedy','Action'])].sort_values(by=['rating'],ascending=False)
```

Best rated movie, with least amount of revenue generated (using quantiles, top 1% of rating, top 25% of revenue):

```python
df1[
    (df1['rating'] > df1['rating'].quantile(0.99))
    & (df1['rev'] < df1['rev'].quantile(0.25))
]
```
